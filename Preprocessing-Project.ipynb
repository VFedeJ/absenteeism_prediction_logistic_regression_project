### Columna Reason for Absence

Esta columna nos va a indicar la razon del absentismo el determinado dia en el registro. Vemos que las razones son numericas por lo que interpretamos que son una referencia a otra cosa. En el material del proyecto podemos observar cada una de estas causas y su numero correspondiente.




```python
df['Reason for Absence'].max()
```

```python
df['Reason for Absence'].unique()
```

Como estos son valores categoricos (ya que se asignan correspondiente a cada causa) debemos obtener los dummies de cada una de ellas

```python
razones=pd.get_dummies(df['Reason for Absence'],drop_first=True)
```

```python
razones
```

```python
#Creamos una variable para chequear que se asigno a cada variable un valor
razones['chequeo']=razones.sum(axis=1)
```

```python
razones
```

```python
razones['chequeo'].unique()
```

```python
#Eliminamos rapidamente para no alterar nuestro dataset
razones=razones.drop(['chequeo'],axis=1)
```

```python
razones
```

Una vez que tenemos los dummies, eliminamos nuestra feature original para reemplazarla

```python
df=df.drop(['Reason for Absence'],axis=1)
```

```python
df
```

```python
razones.loc[:,15:17].max(axis=1)
```

Teniendo tantas variables categoricas puede ser que nuestro algoritmo genere overfitting o procese de manera muy lenta, por ende para simplificarlo lo que haremos es agrupar estas razones de a grupos. Estos grupos seran de acuerdo a las causas parecidas. 
La forma de asignacion sera la siguiente:


```python
grazon1=razones.loc[:,1:14].max(axis=1)
grazon2=razones.loc[:,15:17].max(axis=1)
grazon3=razones.loc[:,18:21].max(axis=1)
grazon4=razones.loc[:,22:].max(axis=1)
```

Agrupando tomando el valor maximo (1) si es que esa causa aplica a cada registro


```python
grazon2
```

Concatenamos a nuestro Dataframe con el que trabajamos los distintos grupos de casusas

```python
df=pd.concat([df,grazon1,grazon2,grazon3,grazon4],axis=1)
```

```python
df
```
